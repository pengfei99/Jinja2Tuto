# Jinja2 template demo

In this notebook, we will show how to use jinja2 template to generate a report in html format.

```python
from jinja2 import Environment, FileSystemLoader
from pathlib import Path
```

```python
template_path = "C:/Users/pliu/Documents/git/Jinja2Tuto/jinja2Tuto/templates/"
template1_name = "hello.html"
report_template_name = "budget_report.html"
```

```python
data1 = {
    "user": {
        "name": "Dalia",
        "roles": ["admin", "editor","developer"]
    }
}
```

```python

env = Environment(loader=FileSystemLoader(template_path))
template1 = env.get_template(template1_name)



```

```python
html1 = template1.render(data1)
print(html1)
```

```python
report_dir = Path("../../data/tmp/out")

report_dir.mkdir(parents=True, exist_ok=True)

output_path1 = report_dir / "pengfei_demo.html"

#
output_path1.write_text(html1, encoding="utf-8")
```

```python
def generate_amount(in_val: int)->int:
    return in_val*10000
```

```python

report_data = {
    "organization": {
        "name": "CASD",
        "department": "Projet Management"
    },
    "report": {
        "period": "2026",
        "author": "Dalia",
        "date": "2025-12-15",
        "summary": "This report provides an estimated budget for planned activities."
    },
    "budget": {
        "rows": [
            {"category": "Infrastructure", "description": "Servers and storage", "amount": generate_amount(12)},
            {"category": "Software", "description": "Licenses and subscriptions", "amount": generate_amount(45)},
            {"category": "Staffing", "description": "Contractors and training", "amount": generate_amount(8)},
            {"category": "Hobby", "description": "for the pleasure of employee", "amount": generate_amount(999)},
        ],
        "total": 245000
    }
}
```

```python
report_template = env.get_template(report_template_name)

html2 = report_template.render(report_data)
print(html2)
```

```python
output_path2 = report_dir / "budget_report.html"

#
output_path2.write_text(html2, encoding="utf-8")
```
