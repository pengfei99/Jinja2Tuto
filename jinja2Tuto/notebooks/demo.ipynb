# Jinja2 template demo

In this notebook, we will show how to use jinja2 template to generate a report in html format.

```python
from jinja2 import Environment, FileSystemLoader
from pathlib import Path
```

```python
template_path = "C:/Users/pliu/Documents/git/Jinja2Tuto/jinja2Tuto/templates/"
template_name = "hello.html"
```

```python
data = {
    "user": {
        "name": "Dalia",
        "roles": ["admin", "editor","developer"]
    }
}
```

```python

env = Environment(loader=FileSystemLoader(template_path))
template = env.get_template(template_name)



```

```python
html = template.render(data)
print(html)
```

```python

output_path = Path("C:/Users/Public/Documents") / "hello_pengfei.html"
output_path.mkdir(parents=True, exist_ok=True)
output_path.write_text(html, encoding="utf-8")
```
